```python
# Example of a for loop
a = ['Red', 3.45, 'a', [1,2,3]]

for v in a:
    print(v)
```

```python
# Example of a for loop wih enumerate
a = ['Red', 3.45, 'a', [1,2,3]]

for i, v in enumerate(a):
    print(i, ':', v)
```

```python
# Example of a for loop using the range operator
for i in range(0,11):
    print(i, ':', i**2)
```

```python
# Example of a while loop
a = 0

while a <= 10:
    print(a, ':', a**2)
    a+=1
```

```python
# Iterate keys in a dictionary
dic = {'Red':30, -2.5:'White', 'Orange':[50, 60,70], 5:{'a':1, 'b':-2, 'c':10}}

for v in dic:
    print(v, ':', dic[v])
```

```python
# Iterate keys in a dictionary
dic = {'Red':30, -2.5:'White', 'Orange':[50, 60,70], 5:{'a':1, 'b':-2, 'c':10}}

for i, v in enumerate(dic):
    print('[' + str(i) + ']', v, ':', dic[v])
```
